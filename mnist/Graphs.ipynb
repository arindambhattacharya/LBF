```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
x_axis = np.array([0, 1])
width = 0.3
plt.bar(x_axis - width, [np.mean(ialbf_init_fpr), np.mean(ialbf_insert_fpr_all)], 
        yerr = [np.std(ialbf_init_fpr), np.std(ialbf_insert_fpr_all)], capsize=0.1,
        width=width, label='IA-LBF')
plt.bar(x_axis , [np.mean(calbf1_init_fpr), np.mean(calbf1_insert_fpr_all)], 
        yerr = [np.std(calbf1_init_fpr), np.std(calbf1_insert_fpr_all)], capsize=0.1,
        width=width, label='CA-LBF I')
plt.bar(x_axis + width, [np.mean(calbf2_init_fpr), np.mean(calbf2_insert_fpr_all)], 
        yerr = [np.std(calbf2_init_fpr), np.std(calbf2_insert_fpr_all)], capsize=0.1,
        width=width, label='CA-LBF II')
plt.xticks([0, 1], ['Initial', 'Final'])
plt.ylim([0, 0.1])
plt.ylabel('False Positive Rate')
plt.title('False positive rates for MNIST (all data)')
plt.legend()
plt.savefig('plots/mnist_fpr.png')
```
